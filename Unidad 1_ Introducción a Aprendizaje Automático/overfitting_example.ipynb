```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Datos de ejemplo
X_train = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y_train = np.array([2, 3, 4, 2, 6])

X_test = np.array([6, 7]).reshape(-1, 1)
y_test = np.array([4, 5])

plt.scatter(X_train, y_train)
plt.title('datos de ejemplo')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Creo el modelo de regresión lineal. Lo fiteo a los datos de entrenamiento.
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# ¿Como se ajustó mi modelo para los datos de entrenamiento?
y_train_pred = model.predict(X_train)
plt.plot(X_train, y_train_pred, label='ajustse de la regresión lineal a los datos', color='r')
plt.scatter(X_train, y_train, label='etiquetas reales')
plt.legend()
plt.show()
```

```python
# ¿Ese ajuste es bueno o malo?
from sklearn.metrics import mean_squared_error, r2_score

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_train, y_train_pred)

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y_train, y_train_pred)

# Imprimir las métricas de evaluación
print(f'Error cuadrático medio (MSE): {round(mse, 2)}')
print(f'Coeficiente de determinación (R^2): {round(r2,2)}')
```

```python
# Realizar predicciones para nuevos valores de x
y_pred = model.predict(X_test)
```

```python
# Coeficientes de la recta de regresión
pendiente = model.coef_[0]
intercepto = model.intercept_
round(pendiente,3), round(intercepto,3)
```

```python
# Visualización de los datos y la recta de regresión
plt.scatter(X_train, y_train, color='blue', label='Datos de entrenamiento')
plt.scatter(X_test, y_test, color='orange', label='Datos de prueba')

plt.plot(np.concatenate([X_train, X_test]), np.concatenate([y_train_pred, y_pred]), color='red', label='Recta de regresión')

plt.scatter(X_test, y_pred, color='green', label='Predicciones')


plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Regresión lineal en Python')
plt.grid(True)
plt.show()
```

```python
# Imprimir los coeficientes de la recta de regresión
print(f'Coeficiente (pendiente): {pendiente}')
print(f'Intercepto: {intercepto}')
```

```python
# Sobre el conjunto de test

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y_test, y_pred)

# Imprimir las métricas de evaluación
print(f'Error cuadrático medio (MSE): {round(mse, 2)}')
print(f'Coeficiente de determinación (R^2): {round(r2,2)}')
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
model_tree = DecisionTreeRegressor()
model_tree.fit(X_train, y_train)
```

```python
# ¿Como se ajustó mi modelo para los datos de entrenamiento?
y_train_pred_tree = model_tree.predict(X_train)
plt.plot(X_train, y_train_pred_tree, label='ajuste del árbol', color='r')
plt.scatter(X_train, y_train, label='etiquetas reales')
plt.legend()
plt.show()
```

```python
# Visualizar el árbol de decisión
plt.figure(figsize=(8, 8))
plot_tree(model_tree, feature_names=['X'], filled=True, rounded=True)
plt.title('Árbol de decisión')
plt.show()
```

```python
# Predicciones para los nuevos valores de X
y_pred_tree = model_tree.predict(X_test)

# Visualización de los datos, las predicciones y las regiones del árbol
plt.scatter(X_train, y_train, color='blue', label='Datos de entrenamiento')
plt.scatter(X_test, y_test, color='orange', label='Datos de prueba')

plt.plot(X_train, y_train_pred_tree, color='red', label='Ajuste del árbol')
plt.plot(np.concatenate([X_train, X_test]), np.concatenate([y_train_pred_tree, y_pred_tree]), color='green', label='Regiones del árbol')
plt.scatter(X_test, y_pred_tree, color='green', label='Predicciones')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Árbol de decisión - Regiones')
plt.grid(True)
plt.show()
```

```python
# Calcular el error cuadrático medio (MSE)
mse_tree = mean_squared_error(y_train, y_train_pred_tree)

# Calcular el coeficiente de determinación (R^2)
r2_tree = r2_score(y_train, y_train_pred_tree)

# Imprimir las métricas de evaluación
print(f'Error cuadrático medio (MSE): {mse_tree}')
print(f'Coeficiente de determinación (R^2): {r2_tree}')
```

```python
# Sobre el conjunto de test

y_pred = model_tree.predict(X_test)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y_test, y_pred)

# Imprimir las métricas de evaluación
print(f'Error cuadrático medio (MSE): {round(mse, 2)}')
print(f'Coeficiente de determinación (R^2): {round(r2,2)}')
```

```python
# Realizar predicciones para nuevos valores de x
x_new_tree = np.linspace(0, 7, 100).reshape(-1, 1)
y_pred_tree = model_tree.predict(x_new_tree)
```

```python
# Visualización de los datos y la recta de regresión
plt.scatter(X_train, y_train, color='blue', label='Datos de entrenamiento')
plt.plot(x_new_tree, y_pred_tree, color='red', label='Recta de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Árbol de Decisión Regresor')
plt.grid(True)
plt.show()
```
